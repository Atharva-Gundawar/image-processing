```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images11.jpg')
img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
detector = cv2.CascadeClassifier('Datasets/FaceCascade/templatedata.xml')
```

```python
faces = detector.detectMultiScale(img,1.05)
```

```python
for f in faces:
    x,y,w,h = f
    green = (0,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img)
```

```python
camera = cv2.VideoCapture(0)
while True:
    b,img = camera.read()
    if b == False:
        print("Some error")
        continue
    
    
    detector = cv2.CascadeClassifier('Datasets/FaceCascade/templatedata.xml')
    faces = detector.detectMultiScale(img,1.05)
#    no_of_people,xyz=faces.shape
#    print("numberof faces" , no_of_people)
    for f in faces:
        x,y,w,h = f
        green = (0,255,0)
        cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    
        cv2.imshow("IMAGE",img)

    k = cv2.waitKey(1) & 0xFF
    if k == ord('q'): 
        break
    
    
cv2.destroyAllWindows() 
camera.release()    
```
